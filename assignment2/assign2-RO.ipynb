## Load Necessary Packages

```python
import numpy as np
import pandas as pd
import mlrose
import timeit
import matplotlib.pyplot as plt
import random

```

## Problem 1: 4 Peaks

```python
# Define fitness function and problem
fitness = mlrose.FourPeaks()
problem = mlrose.DiscreteOpt(length=100, fitness_fn=fitness, maximize=True)

start_time = timeit.default_timer()
rhc_best_state, rhc_best_fitness, rhc_fitness_curve = mlrose.random_hill_climb(
    problem, max_attempts=100, max_iters=500, curve=True, random_state=50
)
end_time = timeit.default_timer()
rhc_time = end_time - start_time
print("RHC Time: {:.10f}".format(rhc_time))

# SA
start_time = timeit.default_timer()

sa_best_state, sa_best_fitness, sa_fitness_curve = mlrose.simulated_annealing(
    problem, max_attempts=100, max_iters=500, curve=True, random_state=50
)
end_time = timeit.default_timer()
sa_time = end_time - start_time
print("SA Time: {:.10f}".format(sa_time))

# GA
start_time = timeit.default_timer()
ga_best_state, ga_best_fitness, ga_fitness_curve = mlrose.genetic_alg(
    problem, max_attempts=100, max_iters=500, curve=True, random_state=50
)
end_time = timeit.default_timer()
ga_time = end_time - start_time
print("GA Time: {:.10f}".format(ga_time))

# MIMIC
start_time = timeit.default_timer()
mimic_best_state, mimic_best_fitness, mimic_fitness_curve = mlrose.mimic(
    problem, max_attempts=100, max_iters=500, curve=True, random_state=50
)
end_time = timeit.default_timer()
mimic_time = end_time - start_time
print("MIMIC Time: {:.10f}".format(mimic_time))

while (len(rhc_fitness_curve) != 500):
    rhc_fitness_curve = np.append(rhc_fitness_curve, rhc_fitness_curve[-1])
while (len(sa_fitness_curve) != 500):
    sa_fitness_curve = np.append(sa_fitness_curve, sa_fitness_curve[-1])
while (len(ga_fitness_curve) != 500):
    ga_fitness_curve = np.append(ga_fitness_curve, ga_fitness_curve[-1])
while (len(mimic_fitness_curve) != 500):
    mimic_fitness_curve = np.append(mimic_fitness_curve, mimic_fitness_curve[-1])

# Plot Iterations vs Fitness
iterations = range(1, 501)
plt.plot(iterations, rhc_fitness_curve, label="RHC", color="red")
plt.plot(iterations, sa_fitness_curve, label="SA", color="blue")
plt.plot(iterations, ga_fitness_curve, label="GA", color="green")
plt.plot(iterations, mimic_fitness_curve, label="MIMIC", color="purple")
plt.legend(loc="best")
plt.xlabel("Iterations")
plt.ylabel("Fitness")
plt.show()

# Plot Time Table
# https://www.geeksforgeeks.org/creating-a-pandas-dataframe-using-list-of-tuples/
data = [
    ("RHC", round(rhc_time, 5)),
    ("SA", round(sa_time, 5)),
    ("GA", round(ga_time, 5)),
    ("MIMIC", round(mimic_time, 5)),
]

df = pd.DataFrame(data, columns=["RO Algorithm", "Time"])
df.style
```
